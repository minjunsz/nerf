```python
from pathlib import Path
import sys
module_path = "/home/minjunsz/codes/nerf/src"
sys.path.append(module_path)
from dataset.lego import LegoDataset

data_path = Path.cwd() / '../data/nerf_synthetic/lego'
train_dataset = LegoDataset(data_path, "train", half_res=True)
```

```python
import matplotlib.pyplot as plt
viewdata = train_dataset[0]
img = viewdata.imgs
```

```python
import torch

height, width = 400, 400
dH = int(height // 2 * 0.5)
dW = int(width // 2 * 0.5)
coords = torch.stack(
                torch.meshgrid(
                    torch.linspace(height // 2 - dH, height // 2 + dH - 1, 2 * dH),
                    torch.linspace(width // 2 - dW, width // 2 + dW - 1, 2 * dW),
                    indexing="ij",
                ),
                dim=-1,
            )
```

```python
coords = coords.reshape(-1, 2).long()
```

```python
target_pixels = img[coords[:, 0], coords[:, 1]]
```

```python
plt.imshow(target_pixels.reshape(200,200,3).numpy())
```
